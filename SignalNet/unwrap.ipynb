```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
i = np.arange(0, 1024)
x = np.pi + np.pi*np.sin(2*np.pi*0.05*i) + 0.001*i
x += np.random.normal(0, 0.4, 1024)
plt.plot(x)

# Linear least squares (linear solve)
# A = np.vstack([i, np.ones(len(i))]).T
# m, c = np.linalg.lstsq(A, x, rcond=None)[0]
# plt.plot(i, m*i + c, 'r')

# Linear least squares (with formula)
m = (np.mean(i*x) - np.mean(i)*np.mean(x)) / (np.mean(i**2) - np.mean(i)**2)
c = np.mean(x) - m*np.mean(i)
print(m, c)
plt.plot(i, m*i + c, 'r')
```

```python
for j in range(1, 10):
    x[x > 2 * np.pi] -= 2*np.pi
plt.plot(x[:100])
```

```python
x = np.unwrap(x)
x = np.convolve(x, np.ones(5)/5, mode='same')
plt.plot(x[100:200])
```

```python
diff1 = np.convolve(x, [-1, 1],mode='same')
plt.plot(diff1[100:200])
diff2 = np.convolve(x, [1/12, -2/3, 0, 2/3, -1/12],mode='same')
plt.plot(diff2[100:200])
```
